```python
!pip install word2number
import pandas as pd
import numpy as np
from word2number import w2n
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/hiring.csv')
df.shape
```

```python
df.isnull().sum()
```

```python
df["test_score(out of 10)"].mean()
```

```python
df["test_score(out of 10)"]=df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].mean())
```

```python
df["experience"].fillna("zero",inplace=True)
```

```python
!pip install word2number
from word2number import w2n
df["experience"]=df["experience"].astype(str).apply(w2n.word_to_num).astype(int)
```

```python
df["experience"].mean()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

lr = LinearRegression()
lr.fit(X, y)

y_pred = lr.predict(X)
y_pred
```

```python
exp=float(input("Experience : "))
test_score=float(input("Test Score : "))
interview_score=float(input("Interview Score : "))
user_pred=lr.predict([[exp,test_score,interview_score]])
print("Predicted Salary : $",user_pred)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index,y,label="Actual Salary",marker='o')
plt.plot(df.index,y_pred,label="Model prediction",marker='x')
plt.scatter(len(df),user_pred,color='red',s=100,label="Your Predicted salary")
plt.xlabel("Data point index")
plt.ylabel("Salary($)")
plt.legend()
plt.grid(True)
plt.show()



```
